# To check whether alkaline phosphatase indicates Hypophosphatasia or liver disorder?

```python
#patient.head()
```

```python
Hypophosphatasia=[]
liver_disorder=[]
for i in patient['Alkaline_Phosphotase']:
    if i<=130:
        Hypophosphatasia.append("Yes")
    else:
         Hypophosphatasia.append("No")

    if i>130:
        liver_disorder.append("Yes")
    else:
         liver_disorder.append("No")

patient['Hypophosphatasia']=Hypophosphatasia
patient['liver_disorder']=liver_disorder
patient
```

```python
#(patient['Alkaline_Phosphotase']<=130)
```

# Inference: If alkaline phosphatase is low below 130 means indicates Hypophosphatasia(affects bone&teeth) and high means above 130 indicates liver disorder patients
